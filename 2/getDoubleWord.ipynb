```python
import pandas as pd
df = pd.read_csv("./data.csv")
df.head()
```

```python
df = df[df["y-kien"].notnull()]
df = df[df["goc-nhin"].notnull()]
df.head()
```

```python
df = df.drop(['STT', 'Mã SV', "Họ và tên"], axis=1)
```

```python
num_columns = df.columns.shape[0]
col_names = df.columns.values.tolist()
df.loc[:, 'merged'] = ""
for i in range(num_columns):
    df.loc[:, 'merged'] = df.loc[:, 'merged'] + df[col_names[i]]
```

```python
list = df.loc[:, 'merged']
df_trans = pd.DataFrame(list).reset_index()
df_trans.head()
```

```python
text = df_trans.groupby("merged")["merged"].apply(' '.join)
```

```python
import re
from string import punctuation
from collections import Counter
title = df.columns.tolist()
t = ''
for i in title:
    t += df.loc[20][i]
sentence = str(text).replace("\\n", " ").translate(punctuation).lower()
words = re.findall(r'\w+', sentence)
two_words = [' '.join(ws) for ws in zip(words, words[1:])]
wordscount = {w:f for w, f in Counter(two_words).most_common() if f > 1}
wordscount
```
