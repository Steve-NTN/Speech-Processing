```python
import pandas as pd
df = pd.read_csv("./data.csv")
df.head()
```

```python
df = df[df["y-kien"].notnull()]
df = df[df["goc-nhin"].notnull()]
df
```

```python
dfd = df.drop(['STT', 'Mã SV', "Họ và tên"], axis=1)
f = open("./top5wordsAudio/yteII.txt", "w", encoding="utf-8")
title = dfd.columns.tolist()

for i in title:
    s = dfd.loc[25][i]
    listStr = s.split("\n")
    getIndex = [listStr[i-1] for i in range(len(listStr)) if "y tế" in listStr[i]]
    if i == 'y-kien': f.write(i) 
    else: f.write("\n"+i) 
    for j in getIndex: f.write("\n" + j)
f.close()
```
