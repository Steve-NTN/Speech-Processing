```python
import pandas as pd
df = pd.read_csv("./data.csv")
df.head()
```

```python
df = df[df["y-kien"].notnull()]
df = df[df["goc-nhin"].notnull()]
df.loc[25]['khoa-hoc']
```

```python
#df = df.drop(['STT', 'Mã SV', "Họ và tên"], axis=1)
f = open("indexAudio.txt", "w")
title = df.columns.tolist()
for i in title:
    s = df.loc[20][i]
    listStr = s.split("\n")
    getIndex = [listStr[i-1] for i in range(len(listStr)) if "của" in listStr[i]]
    if i == 'y-kien': f.write(i) 
    else: f.write("\n"+i) 
    for i in getIndex: f.write("\n" + i)
f.close()
```
